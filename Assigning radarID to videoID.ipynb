```python
import pandas as pd
import numpy as np
```

```python
fused_data1 = pd.read_csv('velocity_and_distance_fused.csv')
fused_data2 = pd.read_csv('distancefused.csv')
fused_data = fused_data1
print(fused_data1)
radar = pd.read_csv('radar_data.csv')
```

```python
frradar = np.unique(fused_data.radarID.values)
start  = fused_data.tstamp[0]
fr_lenradar = frradar.shape[0]
i = 0 
j = 0 
assigneditems = pd.DataFrame({'radarID': [],'videoID': [], 'vehicle': [], 'tstamp': []})
usedvideoIDs = []
while i < fr_lenradar:
    fusedradarIDs = fused_data.loc[(fused_data['radarID'] == frradar[i])].values
    VideoIDs = []
    j = 0 
    for item in fusedradarIDs:
        time = (fusedvideoIDs[0][4]-start)/60000
        place = fusedvideoIDs[0][1]
        if fusedradarIDs[j][7] != ' ':
            VideoIDs.append(fusedradarIDs[j][7])
            if (fusedradarIDs[j][1] - 60.16055) < (place -60.16055):
                time = (fusedradarIDs[j][4]-start)/60000
                place = fusedradarIDs[j][1]
        j = j + 1
    if len(VideoIDs) > 5:
        videoid = max(set(VideoIDs), key = VideoIDs.count)
        count = VideoIDs.count(videoid)
        fusedvideoIDs = fused_data.loc[(fused_data['videoID'] == videoid)].values
        radarIDs = [] 
        j = 0 
        for item in fusedvideoIDs:
            radarIDs.append(fusedvideoIDs[j][6])
            j = j + 1
        radarid = max(set(radarIDs), key = radarIDs.count)
        if frradar[i] == radarid:
            assigneditems2 = pd.DataFrame({'radarID': [frradar[i]],'videoID': [videoid], 'vehicle': [fusedvideoIDs[0][3]], 'tstamp':[time]})
            usedvideoIDs.append(videoid)
            assigneditems = assigneditems.append(assigneditems2)
    i = i + 1 
assigneditems.to_csv('assigned_radar+video.csv')
print(assigneditems)    
```
