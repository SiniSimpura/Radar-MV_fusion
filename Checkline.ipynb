```python
radar = pd.read_csv('radarwithacceleration.csv')
i = 0 # i kertoo mitä havaintoriviä käydään läpi radardatasta
frradar = np.unique(radar.tstamp.values)
fr_lenradar = frradar.shape[0]
radar_checkline = pd.DataFrame({'tstamp': [], 'v_n': [],'v_e': [], 'a_n': [],'a_e': [], 'radarID':[],'lane1':[],'lane2':[],'type':[]})
lane4 = 0
lane3 = 0
lane2 = 0 
lane1 = 0 
while i < (fr_lenradar): 
    radardetects = radar.loc[(radar['tstamp'] == frradar[i])].values
    j = 0
    while j < len(radardetects):
        if radardetects[j][7]> 8:
            vehicle = 'truck'
        elif radardetects[j][7] > 4:
            vehicle = 'car'
        else:
            vehicle = 'other'
        if radardetects[j][1] > 60.16055 and radardetects[j][1] < 60.16065 and radardetects[j][8] not in radar_checkline.radarID.values and radardetects[j][12] > 50:
            print(radardetects[j][2])
            if radardetects[j][2]>=24.92215 and radardetects[j][2] < 24.9222: 
                radar_checkline2 = pd.DataFrame({'tstamp': [radardetects[j][13]], 'v_n': [radardetects[j][3]],'v_e': [radardetects[j][4]], 'a_n': [radardetects[j][5]],'a_e': [radardetects[j][6]], 'radarID':[ radardetects[j][8]],'lane1':[4],'lane2':[radardetects[j][9]],'type':[vehicle]})
                radar_checkline = radar_checkline.append(radar_checkline2)
                lane4 = lane4 +1
            if radardetects[j][2] < 24.92215 and radardetects[j][2] >= 24.922:
                radar_checkline2 = pd.DataFrame({'tstamp': [radardetects[j][13]], 'v_n': [radardetects[j][3]],'v_e': [radardetects[j][4]], 'a_n': [radardetects[j][5]],'a_e': [radardetects[j][6]], 'radarID':[ radardetects[j][8]],'lane1':[3],'lane2':[radardetects[j][9]],'type':[vehicle]})
                radar_checkline = radar_checkline.append(radar_checkline2)
                lane3 = lane3 +1
            if radardetects[j][2] < 24.922 and radardetects[j][2] >= 24.92188:
                radar_checkline2 = pd.DataFrame({'tstamp': [radardetects[j][13]], 'v_n': [radardetects[j][3]],'v_e': [radardetects[j][4]], 'a_n': [radardetects[j][5]],'a_e': [radardetects[j][6]], 'radarID':[ radardetects[j][8]],'lane1':[1],'lane2':[radardetects[j][9]],'type':[vehicle]})
                radar_checkline = radar_checkline.append(radar_checkline2)
                lane2 = lane2 +1
            if radardetects[j][2] < 24.92188 and radardetects[j][2] > 24.92175:
                radar_checkline2 = pd.DataFrame({'tstamp': [radardetects[j][13]], 'v_n': [radardetects[j][3]],'v_e': [radardetects[j][4]], 'a_n': [radardetects[j][5]],'a_e': [radardetects[j][6]], 'radarID':[ radardetects[j][8]],'lane1':[0],'lane2':[radardetects[j][9]],'type':[vehicle]})
                radar_checkline = radar_checkline.append(radar_checkline2)
                lane1 = lane1 +1             
        j = j +1
    i = i + 1
print(lane1, lane2, lane3, lane4)
print(radar_checkline)
radar_checkline.to_csv('radar_checkline.csv')
```
